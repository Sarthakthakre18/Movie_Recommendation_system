```python
import ast
def extracted_feature(text):

    L=[]
    text = ast.literal_eval(text)            # convert this : "[{'id':28,'name':'Science Fiction'}]"
    for i in text:                                  # to this : ['ScienceFiction']
    

        L.append(i["name"].replace(" ",""))

     


    return L

        
```

```python
movies["genres"] = movies["genres"].apply(extracted_feature)

movies["keywords"] = movies["keywords"].apply(extracted_feature)

movies["cast"] = movies["cast"].apply(extracted_feature)
```

```python
def extracted_crew(text):

    L = []

    text = ast.literal_eval(text)

    for i in text:

        if i["job"] == "Director":

          L.append(i["name"].replace(" ",""))

        

    return L
```

```python
movies["crew"] = movies["crew"].apply(extracted_crew)
```

```python
movies.head()
```

```python
def split_overview(text):

    return text.split()
```

```python
movies["overview"] = movies["overview"].apply(split_overview)
```

```python
movies.head()
```

```python
movies["tags"] = (
    movies["overview"]
    + movies["genres"]
    + movies["keywords"]
    + movies["cast"]
    + movies["crew"]
)
```

```python
movies.iloc[0]["tags"]
```

```python
movies["tags"] = movies["tags"].apply(lambda x: " ".join(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(
    max_features=5000,
    stop_words="english"
)

vectors = cv.fit_transform(movies["tags"]).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similaritys = cosine_similarity(vectors)
```

```python
similaritys.shape
```

```python
movie_list = list(enumerate(similaritys[0]))
```

```python
sorted_movies = sorted(
    movie_list,
    reverse=True,
    key=lambda x: x[1]
)
```

```python
def recommend(movie):

    movie_index = movies[movies["title"] == movie].index[0]

    distance = similaritys[movie_index]

    movie_list = list(enumerate(distance))

    sorted_movies = sorted(
        movie_list,
        reverse=True,
        key=lambda x: x[1]
    )

    for i in sorted_movies[1:6]:

        print(movies.iloc[i[0]].title)
```

```python
recommend("Lifeforce")
```

```python
from nltk.stem.porter import PorterStemmer # Stemming converts different forms of a word into a common base word.
```

```python
ps = PorterStemmer()
```

```python
ps.stem("loving")
```

```python
def stem(text):

    L = []

    for i in text.split():

        L.append(ps.stem(i))

    return " ".join(L)
```

```python
movies["tags"] = movies["tags"].apply(stem)
```

```python
cv = CountVectorizer(
    max_features=5000,
    stop_words="english"
)
```

```python
vectors = cv.fit_transform(movies["tags"]).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
recommend("Avatar")
```

```python
recommend("Batman Begins")
```

```python
recommend("Avatar")
recommend("The Dark Knight")
recommend("Interstellar")
recommend("Titanic")
recommend("Iron Man")
```
